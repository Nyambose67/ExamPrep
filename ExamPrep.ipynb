```python
import numpy as np

def bubble_sort(arr):
    n = len(arr)
    # Traverse through all array elements
    for i in range(n):
        # Last i elements are already sorted
        for j in range(0, n-i-1):
            # Swap if the element found is greater than the next element
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]


# Extract the column to be sorted
column_to_sort = 'Proportion of urban population living in slums or informal settlements (%) (a)'

# Convert the column to numeric, forcing errors to NaN
df[column_to_sort] = pd.to_numeric(df[column_to_sort], errors='coerce')

# Drop NaN values from the column before sorting
data_to_sort = df[column_to_sort].dropna().values

# Perform Bubble Sort
bubble_sort(data_to_sort)

# Add the sorted data back to the DataFrame
df_sorted = df.copy()

# Create a new column for sorted data (keeping original rows intact)
df_sorted[column_to_sort + ' (sorted)'] = np.nan
df_sorted.loc[df[column_to_sort].dropna().index, column_to_sort + ' (sorted)'] = data_to_sort

# Display the first few rows of the sorted DataFrame
print("\nSorted DataFrame (ascending order):")
print(df_sorted.head())
```
